Data Exploration Using Seaborn

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

Dataset Tips from seaborn

```python
tips=sns.load_dataset('tips')
df=pd.DataFrame(tips)
df.head()
```

```python
df.shape
```

```python
sns.set_context('paper',font_scale=1)
sns.distplot(df['tip'])
```

```python
corr_matrix=df.corr()
sns.heatmap(data=corr_matrix, annot=True)
```

```python
sns.relplot(x='total_bill',y='tip',data=df,col='time',hue='smoker', size='size')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,col='time',hue='smoker')
```

```python
sns.catplot(x='day',y='tip',data=df,kind='swarm')
```

```python
sns.catplot(x='day',y='tip',data=df,kind='violin')
```

```python
sns.catplot(x='day',y='tip',data=df,kind='bar')
```
