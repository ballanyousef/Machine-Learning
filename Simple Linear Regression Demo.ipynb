#First Simple Linear Regression Program

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##Find Coefficient values

```python
def find_coef(x,y):
    n=np.size(x)
    
    #mean values of vector x and y
    mean_x,mean_y=np.mean(x), np.mean(y)
    
    #calculate cross deviations
    Sxy=np.sum(x*y)-n*mean_x*mean_y
    Sxx=np.sum(x*x)-n*mean_x*mean_x
    
    #coefficients
    b1=Sxy/Sxx
    b0=mean_y-b1*mean_x
    
    return(b0,b1)
```

##Plot regression line

```python
def plt_regression_line(x,y,b):
    plt.scatter(x,y,color="m", marker="o", s=30)
    
    #predict y
    y_pred=b[0]+b[1]*x
    
    #plot the regression line
    plt.plot(x,y_pred,color="g")
    
    #print label
    plt.xlabel('x')
    plt.ylabel('y')
    
    #show plot
    plt.show()
```

##main function

```python
def main():
    tips=sns.load_dataset('tips')
    df=pd.DataFrame(tips)
    print(df.head())
    
    #observations
    x=df['total_bill']
    y=df['tip']
    
    #estimate coeff function
    b=find_coef(x,y)
    
    #plot regression line
    plt_regression_line(x,y,b)
    
if __name__=="__main__":
    main()
```
